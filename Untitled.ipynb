```python
#!pip install selenium
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
chromedriver = 'C:/chromedriver_win32/chromedriver'
driver = webdriver.Chrome(chromedriver)
```

```python
name = input('아이디: ')
driver.get('https://op.gg')
a = driver.find_element_by_name('userName')
a.send_keys(name)
a.submit()
driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div/div[5]/div[2]/div[2]/div/div[1]/div/ul/li[2]/a').click()
time.sleep(3)


for i in range(4,10):
    driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div/div[5]/div[2]/div[2]/div/div[2]/div[{}]/a'.format(i)).click()
    time.sleep(5)

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
```

```python
cham =[]
result =[]
for i in soup.select('div.Content > div.GameSettingInfo > div.ChampionName > a'):
    cham.append(i.text)
for i in soup.select('div.Content > div.GameStats > div.GameResult'):
    result.append(i.text.replace('\n','').replace('\t',''))
```

```python
data =[cham,result]
df = pd.DataFrame(data).T
```
